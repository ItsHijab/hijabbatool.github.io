```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [25000, 27000, 30000, 28000, 32000, 35000],
    'Profit': [5000, 7000, 8000, 7500, 9000, 9500]
}

df = pd.DataFrame(data)
df.head()
```

```python

df.describe()
```

```python

plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Month'], df['Profit'], marker='s', label='Profit')
plt.title("Monthly Sales & Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount (PKR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

best_month = df.loc[df['Sales'].idxmax(), 'Month']
print(f" Highest Sales Recorded in: {best_month}")
```
